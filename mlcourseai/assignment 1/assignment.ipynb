```python
import pandas as pd
dataframe = pd.read_csv("database/adult/adult.csv", sep=', ')
dataframe.head()
```

1. How many men and women (sex feature) are represented in this dataset?

```python
men = dataframe[dataframe["sex"]=="Male"]
women = dataframe[dataframe["sex"]=="Female"]
```

```python
men['age'].count()
```

```python
women['age'].count()
```

2. What is the average age (age feature) of women?

```python
women['age'].sum()/women['age'].count()
women['age'].mean()
```

3. What is the percentage of German citizens (native-country feature)?

```python
germans = dataframe[dataframe['native-country']=='Germany']
germans['age'].count()/dataframe['age'].count()
```

4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?

```python
more = dataframe[dataframe['salary']=='>50K']
less = dataframe[dataframe['salary']=='<=50K']

print(more['age'].std())
print(more['age'].mean(),'\n')
print(less['age'].std())
print(less['age'].mean())
```

6. Is it true that people who earn more than 50K have at least high school education? (education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate feature)

```python
atLeastHighSchool = ('Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate')
thoseWhoEarn = dataframe[dataframe['salary']=='>50K']

# THOSE WHO EARN >50K, ARE THEY ALL IN AT LEAST HIGH SCHOOL?
thoseWhoEarn['education'].isin(atLeastHighSchool).all()
# there is at least 1 person that has never been to high school, and earn > 50k
```
